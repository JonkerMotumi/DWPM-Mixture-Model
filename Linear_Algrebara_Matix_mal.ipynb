```python
import numpy as np

# Create two random 4x4 matrices
A = np.random.rand(4, 4)
B = np.random.rand(4, 4)
```

```python
def manual_matmul(A, B):
    # Get dimensions
    rows_A, cols_A = A.shape
    rows_B, cols_B = B.shape

    # Initialize the result matrix with zeros
    result = np.zeros((rows_A, cols_B))

    # TODO: Implement triple nested loops here
    for i in range(rows_A):
        ans=0
        for j in range(cols_B):
            ans=0
            for k in range(cols_A):
                x = A[i,k]
                y = B[k,j]
                ans = ans + x*y
                result[i,j]= ans

    return result
```

```python
manual=manual_matmul(A,B)
official = np.matmul(A, B)
print(manual)
print(official)
print(f"Match found: {np.allclose(manual, official)}")
```
