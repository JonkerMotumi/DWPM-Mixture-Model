```python
import numpy as np
a = [[3, 2, 2], [2, 3, -2]]

print("Matrix A:", a)

at= np.transpose(a)

print("Matrix AT:", at)

adotat= np.dot(a,at)

print(adotat)

labda = np.linalg.eigvals(adotat)

print("Eigenvalues:", labda)

print()
```

```python
g= np.dot(at,a)

print(g)

eigenvalues, eigenvectors =np.linalg.eig(g)

print ("Eigenvalues:", eigenvalues)
print ("Eigenvectors:", eigenvectors)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import data, color

# 1. Load a grayscale image
img = color.rgb2gray(data.astronaut())

# 2. Perform SVD
U, S, Vt = np.linalg.svd(img)

# 3. Reconstruct with only 20 "directions" (singular values)
k = 100
compressed_img = U[:, :k] @ np.diag(S[:k]) @ Vt[:k, :]

# 4. Plot
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1); plt.imshow(img, cmap='gray'); plt.title("Original")
plt.subplot(1, 2, 2); plt.imshow(compressed_img, cmap='gray'); plt.title(f"Compressed (k={k})")
plt.show()
```
